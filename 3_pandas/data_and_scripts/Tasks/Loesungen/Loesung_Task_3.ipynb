# Pandas Data Visualization Exercise - Task
This is a small exercise on how to use the built-in visualization capabilities in *Pandas*. For this we use the **df3** CSV. Please visualize the following charts by using Pandas. The imports are given:

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

**Create this scatter plot of b vs. a. Pay attention to the color and size of the dots. Also, try to fit it into the given shape. A refresher on matplotlib commands might help with this.**

```python
df3.plot.scatter(x='a',y='b',c='red',s=50,figsize=(12,3))
```

**Create a histogram of the "a" column.**

```python
df3['a'].plot.hist()
```

**These plots are okay so far. But they do not look consistent. Use the `ggplot` stylesheet to influence the appearance of the following plots. Then recreate the histogram from before and add more subbars.**

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(alpha=0.5,bins=25)
```

**Create a box plot that compares columns a and b.**

```python
df3[['a','b']].plot.box()
```

**Create a KDE plot of the column "d".**

```python
df3['d'].plot.kde()
```

**Find out how we can increase the line width and stroke the line.

*Note: Normally we would not stroke a KDE diagram.

```python
df3['d'].plot.density(lw=5,ls='--')
```

**Create an area plot for all columns. Use only the first 30 rows.**

*Note: For the rows you can use .iloc.*

```python
df3.iloc[0:30].plot.area(alpha=0.4)
```

## Bonus Task!
This could be really hard, which is why it's okay to refer to the solution if you get stuck.

**I'm sure you'll notice that the legend in the previous solution covers up the actual diagram. Can you find a way to display the legend next to the actual diagram?**

*Note: Google search for a good stackoverflow link on this topic helps.*

```python
f = plt.figure()
df3.loc[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
