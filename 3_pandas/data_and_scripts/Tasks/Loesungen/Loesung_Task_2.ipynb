## Exercise

The following instructions should be implemented by using the following data:

### Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt and run %matplotlib inline if you use the Jupyter notebooks. Which command should you use if you don't use the notebooks?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show() for non-notebook users
```

## Exercise 1

**Complete the following steps:**

**Create a diagram object *fig* using plt.figure()**.
**Use add_axes to add an axis to the chart on the canvas at position [0,0,1,1]. Call this axis *ax*.**
**Display (x,y) on the axes and name labels and titles as in the example shown.**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Titel')
```

## Exercise 2

**Create a graph object and add the two axes *ax1* and *ax2*. Place them at [0,0,1,1] and [0.2,0.5,0.2,0.2] respectively.**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

**Now plot (x,y) on both axes. Also, call up your diagram to display it.**

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig # Show figure object
```

## Exercise 3
**Create the diagram you see by adding two axes. Place them at [0,0,1,1] and [0.2,0.5,0.4,0.4].**

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

**Now use x,y, and z arrays to display the following diagram. Pay attention to `xlimits` and `ylimits` in the inserted diagram.

```python
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Blank workspace with 1 x 2 diagrams
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now add (x,y) and (x,z) on the two axes. Play with line width and style.**

```python
axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[1].plot(x,z,color="red", lw=3, ls='-')
fig
```

**Now resize the plots by adding the figsize() argument to plt.subplots(). You can copy your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
