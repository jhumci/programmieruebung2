# Loading data + Different data formats

Importing the used libraries
By "import numpy as np" we can call e.g. numpy functions via _numpy.functionsName_.

````python
import numpy as np
import pandas as pd
import os
````

## Task 1.1
__Load one .csv and one .txt file each from the "ECG_Data" folder and save the file in one variable each of type Numpy Array, List and Pandas Dataframe.__

````python
# define path using a string value
directory =  'EKG_Data/'
file_name = "01_Ruhe.txt"
file_name_csv = "01_Ruhe.csv"


## Open the file using the path and define a certain split character
# The files first column are Values in millivolt, second colum is time in milliseconds
## Write it into a ... 
# List 
list_01 = open(directory + file_name,"r").read().split("\n")
time = []
value = []

for line in list_01:
    _line = line.split("\t")
    value.append(int(_line[0]))
    time.append(int(_line[1]))
    
print("list_01 is of type:" +  str(type(list_01)))
print("time is of type:" +  str(type(time)))
print("value is of type:" +  str(type(value)))

# Numpy Array
array_01_test = np.array(list_01)
print("array_01_test is of type:" +  str(type(array_01_test)))
# The array is still containing the "\t" separator, instead we could use the two lists "value","time" and stack them into one array... input has to be a tupel 
array_01 = np.column_stack((time, value)) 
print("after stacking the two list in one array it is of type:" + str(type(array_01)))

# Pandas dataframe
df_01 = pd.DataFrame(list_01)
print("df_01 is of type:" +  str(type(df_01)))

print("the first 5 rows of the dataframe look like this: ")
print(df_01.head())
````

## Task 1.2

Look up the pandas function __"pd.read_csv"__ and use it, to load the txt. and .csv file into a dataframe. Please also add a header with the column names :"Amplitude in [mV]" ,"Time in [ms]"

````python
# load_csv()
# for txt file
df_ruhe_01 = pd.read_csv(directory + file_name, sep='\t')
print(df_ruhe_01.head())

# for csv file
column_names = ["Amplitude in [mV]" ,"Time in [ms]"]
df_ruhe_01_csv = pd.read_csv(directory + file_name_csv, sep=';', header=0, names= column_names)
df_ruhe_01_csv.head()
````
