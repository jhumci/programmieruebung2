## Task 1.1
__Load one .csv and one .txt file each from the "ECG_Data" folder and save the file in one variable each of type Numpy Array, List and Pandas Dataframe.__

````python
# define path using a string value
directory =  'EKG_Data/'
file_name = "01_Ruhe.txt"
file_name_csv = "01_Ruhe.csv"


## Open the file using the path and define a certain split character
# The files first column are Values in millivolt, second column is time in milliseconds
## Write it into a ... 
# List 
list_01 = open(directory + file_name,"r").read().split("\n")
````

````python
# Looping the list to split the data in two separate lists
time = []
value = []

for line in list_01:
    _line = line.split("\t")
    value.append(int(_line[0]))
    time.append(int(_line[1]))
    
print("list_01 is of type:" +  str(type(list_01)))
print("time is of type:" +  str(type(time)))
print("value is of type:" +  str(type(value)))
````

````python
time
````

````python
# Numpy Array
array_01_test = np.array(list_01)

print("array_01_test is of type:" +  str(type(array_01_test)))
# The array is still containing the "\t" separator, instead we could use the two lists "value","time" and stack them into one array... input has to be a tupel 
array_01 = np.column_stack((time, value)) 
print("after stacking the two list in one array it is of type:" + str(type(array_01)))
````
