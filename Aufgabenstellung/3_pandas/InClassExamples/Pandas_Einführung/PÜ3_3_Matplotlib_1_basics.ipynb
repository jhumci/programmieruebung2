# Matplotlib object oriented

Since you we now successfully started using `matplotlib`, we can take a deeper dive into the object oriented properties of the library. This means that every time we are using the library for a graph, we will initialize a instance of an object and use the methods and attributes of this object.

## Introduction in using object oriented methods

We can start by creating an instance of a graph. Then we will add the axis for the graph.

```python
# create empty graph
af = plt.figure()

# Add axes to plot
axes = af.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (values between 0 and 1)

# Use axes to set label and title
axes.plot(x, y, 'b')

axes.set_xlabel('X label') # Use set_ to begin with a method
axes.set_ylabel('Y label')
axes.set_title('Title name')
```

This can be challenging in the beginning to use `matplotlib` like this but it will help in the future to keep full control over your graphs and code. You could for example add a second axis:

```python
# create empty graph
af = plt.figure()

axes1 = af.add_axes([0.1, 0.1, 0.8, 0.8]) # Main Axis
axes2 = af.add_axes([0.2, 0.5, 0.4, 0.3]) # second Axis

# Bigger Graph using axis 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('x axis 1')
axes1.set_ylabel('y axis 1')
axes1.set_title('Titel bigger diagram')

# second axis
axes2.plot(y, x, 'r')
axes2.set_xlabel('x axis 2')
axes2.set_ylabel('y axis 2')
axes2.set_title('Titel smaller diagram')
```

## subplots()

The plt.subplot() object works like an automatic axis handler. 

Here is one simple use-case:

```python
# Same like above, but now with subplots() 
diag, axes = plt.subplots()

# Now we add more information to the object
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('titel')
```

We can define the number of subplots() and how the are arranged.

```python
# Empty graph with 1 x 2 Subplots
diag, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
# Axes is a array of axis
axes
```

We can iterate through the array:

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('titel')

# Display 
diag
```

One regular problem with `matplotlib` is the overlap of graphs and their axis. We can use ``diag.tight_layout()`` or ``plt.tight_layout()`` to set the position of the axes automatically to ged rid of possible overlaps:

```python
diag, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('titel')

diag    
plt.tight_layout()
```

## Size and DPI

Matplotlib lets you specify the ratio how the graph should be displayed, the DPI, and the size. Therefore we use  `figsize` und `dpi` arguments.

* `figsize` is a tupel with the width and heiht measured in inches
* `dpi` stands for dots-per-inch

For example:

```python
diag = plt.figure(figsize=(8,4), dpi= 400)

ax = diag.add_axes([0,0,1,1])
ax.plot(x,y)
```

The same arguments can also be passed to layout management (such as `subplots`):

```python
diag, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('titel')
```

## Saving graphs

To save graphs please use `savefig` in the `figure` class:

## Legends

Using **label="here comes your label"** as an argument for the graph in combination with the `legend` method (ohne weitere Argumente).

```python
diag = plt.figure()

ax = diag.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

Sometime the label position is not optimal but you can rearange it with the **loc** Argument. Here are some examples:

```python
# some options

ax.legend(loc=1) # right top corner
ax.legend(loc=2) # left top corner
ax.legend(loc=3) # left bottom corner 
ax.legend(loc=4) # right bottom corner

# Most used option
ax.legend(loc=0) # Matplotlib searches for the optimal position and decides on its own
```

## Setting colors, line type and size

Matplotlib gives us many options to specify color, thickness and type of a line (or chart).

There is the MatLab like syntax for this, but also other variants.

### Colors with the MatLab like syntax

With Matplotlib, we can define color and other graphical elements in a variety of ways. The first way would be to define it in a MatLab like syntax. In this one, "b" means blue, "g" means green, etc. The MatLab API also allows us to select the line type. For example, "b.-" means in the diagram means final blue line with dots:

```python
# MatLab Style
diag, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green, dashed line
```

### Colors with the "color =" parameter

Secondly, we can specify colors using the "color =" parameter. For this we use their name or RGB code and optionally an alpha value to set the transparency.

```python
diag, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # Transparency 50%
ax.plot(x, x+2, color="#8B008B")        # RGB Hex Code
ax.plot(x, x+3, color="#FF8C00")        # RGB Hex Code 
```

### Line and highlight style

To change the line width we can use `linewidth` or `lw`. To set the style we can use `linestyle` or `ls`.

```python
diag, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# Possible styles for lines -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# personalised line
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # Format: line length, distance, ...

# Possibles markers: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# Marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Diagram area

We can define the diagram range of the axes with `set_ylim` and `set_xlim` as methods of the axis object. Alternatively, we can use `axis("tight")` to automatically select a "tight" range:

```python
diag, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Standard axis")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Tight axis")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Personalized");
```
