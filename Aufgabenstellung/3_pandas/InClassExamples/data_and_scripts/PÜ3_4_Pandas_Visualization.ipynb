# Pandas built-in Data Visualization
In this lesson we will learn more about the available VIsualization capabilities of *Pandas*. 
It is based on *Matplotlib*, but is available directly in Pandas for easier handling.

Let's start!

### Imports

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

### The data
We have some CSV files with data that we can use as data frames:

```python
df1 = pd.read_csv("df1",index_col=0)
df2 = pd.read_csv("df2")
```

## stylesheets
Matplotlib provides [*Stylesheets*](http://matplotlib.org/gallery.html#style_sheets) that we can use to make our charts look a little better. We will look at some of them as examples. 

We recommend using the stylesheets as they give all diagrams a consistent look. It is also possible to create your own stylesheets to represent a company, for example.

Now let's look at the usage:

*Before using plt.style.use():*.

```python
df1['A'].hist()
```

Now we use the stylesheet.

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

Let's leave it at the *ggplot* stylesheet for now and look at the actual visualization capabilities of *Pandas*.

## Diagram types
There are several built-in visualizations in *Pandas*. Most of them are statistical charts. Here is an overview:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

We can alternatively call df.plot(kind="area").

Now let's take a look at all the plots:

### Area

```python
df2.plot.area(alpha=0.4)
```

### Barplots

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

### Histogramm

```python
df1['A'].plot.hist(bins=50)
```

### Line Plot

```python
df1.plot.line(y='B',lw=1, figsize=(12,3))
```

### Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

We can use except *c* to color the points using a third variable. To do this, we define `cmap` with one of the [Colormaps](http://matplotlib.org/users/colormaps.html).

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

Or we use `s` to change the size (en. scale) of the points. The `s` parameter must be an array. We see here how it works:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

### Box Plots

```python
df2.plot.box() # We can also group by with by= (en. group by)
```

### Hexagonal Bin Plot
A useful alternative to the scatterplot for bivariate data:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

### Kernel Density Estimation (KDE) Plot

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```

And with that we have them all. Hopefully, you will find this variant of visualization easier than complete Matplotlib approach. It provides a good balance of simplicity and control. Also, for many of the diagrams, we can pass more parameters (just like their matplotlib parent diagrams).
